## Import Library and Given Code

```python
import numpy as np

list_ = ['1','2','3','4','5']
array_list = np.array(object = list_)
```

## Q1. Is there any difference in the data type of variables list_ and array_list?

Yes, there is a difference. `list_` is a Python list, while `array_list` is a NumPy array.

### Code to print data types:

```python
print(type(list_))
print(type(array_list))
```

## Q2. Print the data type of each element in list_ and array_list

```python
print('Data types in list_')
for i in list_:
    print(type(i))

print('\nData types in array_list')
for i in array_list:
    print(type(i))
```

## Q3. Considering the change:
`array_list = np.array(object = list_, dtype=int)`

Now the array elements will become integers.

```python
array_list = np.array(object = list_, dtype=int)

print('Data types in list_')
for i in list_:
    print(type(i))

print('\nData types in new array_list')
for i in array_list:
    print(type(i))
```
